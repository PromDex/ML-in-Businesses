```python
'''
Конверсия в генеральной совокупности - это процент или доля индивидов или единиц, обладающих определенным свойством или осуществляющих определенное действие, в рамках всей генеральной совокупности.

Например, в контексте маркетинга и интернет-бизнеса конверсия может относиться к проценту пользователей, которые совершили определенное действие
(например, сделали покупку,заполнили форму, подписались на рассылку и т. д.) из общего числа посетителей или пользователей веб-сайта или приложения.

Конверсия может быть выражена численно или в процентах и является важным показателем эффективности маркетинговых и бизнесовых стратегий.

Измерение и анализ конверсии позволяет оптимизировать процессы, улучшить пользовательский опыт и повысить эффективность деятельности компании.

'''

#Этот код генерирует случайные данные для примера, представляющие генеральную совокупность.
#Затем он расчитывает конверсию в генеральной совокупности путем вычисления среднего значения бинарных данных (1 - успех, 0 - неудача).
#Нулевая гипотеза обычно представляет собой утверждение о равенстве или отсутствии эффекта, альтернативная гипотеза предполагает наличие эффекта или различий.

import numpy as np

# Генерация случайных данных для примера
n = 1000  # размер генеральной совокупности

#p = 0.12  # предполагаемая конверсия в генеральной совокупности 1
p = 0.4   # предполагаемая конверсия в генеральной совокупности 2

# Генерация бинарных данных (1 - успех, 0 - неудача)
population = np.random.choice([0, 1], size=n, p=[1-p, p])

# Расчет конверсии в генеральной совокупности
conversion_rate = np.mean(population)

print("Конверсия в генеральной совокупности:", conversion_rate)
```

Для расчета 95% доверительного интервала для среднего значения используется следующая формула:

Доверительный интервал = (Среднее значение) ± (Критическое значение * (Стандартная ошибка))

Где:

Среднее значение: среднее значение выборки или среднее значение генеральной совокупности (если известно).
Критическое значение: значение, определяющее ширину доверительного интервала и зависит от выбранного уровня доверия. Для 95% доверительного интервала используется значение критического значения, соответствующее 2 стандартным отклонениям.
Стандартная ошибка: стандартное отклонение выборки, поделенное на квадратный корень из объема выборки (если известно), или стандартная ошибка генеральной совокупности.
Пример расчета 95% доверительного интервала для среднего значения:

Предположим, что у нас есть выборка из 100 наблюдений, среднее значение которой равно 50, и стандартное отклонение выборки равно 10.

Стандартная ошибка = 10 / sqrt(100) = 1

Критическое значение для 95% доверительного интервала (двустороннего) составляет приблизительно 1.96.

Доверительный интервал = 50 ± (1.96 * 1) = (47.04, 52.96)

Таким образом, с 95% вероятностью истинное среднее значение будет находиться в интервале от 47.04 до 52.96.

```python
sample_mean = 50 # среднее значение
sample_std = 10 # стандартное отклонение выборки
sample_size = 100 # выборка

standard_error = sample_std / np.sqrt(sample_size)
critical_value = 1.96  # стандартное критическое значение для 95% доверительного интервала

confidence_interval = (sample_mean - critical_value * standard_error, sample_mean + critical_value * standard_error)
print(confidence_interval)
```

```python
'''
Критическое значение определяется на основе выбранного уровня доверия и используется для определения границ доверительного интервала.
Для 95% доверительного интервала, мы используем уровень значимости alpha = 0.05.

Критическое значение можно найти с помощью таблицы значений стандартного нормального распределения (таблицы Z-значений)
или с использованием статистических функций в программном обеспечении, таких как scipy.stats.norm.ppf() в Python.

В случае 95% доверительного интервала, критическое значение равно 1.96, что соответствует двусторонней проверке гипотезы.
Это значение обеспечивает покрытие 95% области под кривой нормального распределения.

В этом примере мы использовали функцию norm.ppf() из библиотеки scipy.stats для расчета критического значения.

Параметр (1 + confidence_level) / 2 используется для двусторонней проверки гипотезы и определения критического значения.

Результат будет примерно равен 1.96, что соответствует критическому значению для 95% доверительного интервала в нормальном распределении.

'''
from scipy.stats import norm

confidence_level = 0.95
critical_value = norm.ppf((1 + confidence_level) / 2)

print("Критическое значение для 95% доверительного интервала:", critical_value)
```

```python
import scipy.stats as st

data = [1, 2, 3, 4, 5]  # пример данных
confidence_interval = st.t.interval(0.95, len(data)-1, loc=np.mean(data), scale=st.sem(data))
print(confidence_interval)
```
